```python
import cv2
import urllib.request
import matplotlib.pyplot as plt
import numpy

url = "https://dcm.ffclrp.usp.br/images/mat.png"

#Open url
url = urllib.request.urlopen(url)

#The read() method returns the specified number of bytes from the file. Default is -1 which means the whole file.
leitura = url.read()

#The bytearray() method returns a bytearray object which is an array of the given bytes.
leitura = bytearray(leitura)

#Convert the input to an array (numpy)
matriz = numpy.asarray(leitura, dtype = "uint8")

#cv2.imdecode() function is used to read image data from a memory cache and convert it into image format.
  #This is generally used for loading the image efficiently from the internet.
matriz = cv2.imdecode(matriz, cv2.IMREAD_COLOR)

#cv2.cvtColor() is an OpenCV function that converts an image from one color space to another.
matriz = cv2.cvtColor(matriz, cv2.COLOR_BGR2RGB)

#Display data as an image
plt.imshow(matriz)
plt.show()


```
